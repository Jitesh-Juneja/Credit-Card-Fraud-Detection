# Metric Used

### ROC-AUC Score

- The ROC-AUC (Receiver Operating Characteristic - Area Under the Curve) score is a commonly used metric to evaluate the effectiveness of binary classifiers.
- It assesses a model's ability to discriminate between positive and negative classes by calculating the area under the ROC curve.
- The ROC-AUC score remains useful for imbalanced datasets because it is not affected by class imbalance.
- Even when positive and negative classes are unbalanced, the ROC-AUC score provides a reliable indication of the model's ability to distinguish between the two classes.

### F1 Score

- The ROC-AUC score doesn't provide detailed insights into a model's performance across individual classes.
- Models may perform differently across classes, especially in imbalanced datasets where one class is significantly smaller than the other.
- While ROC-AUC score might indicate good performance overall, it might overlook poor performance in smaller classes.
- Hence, when evaluating model performance on imbalanced datasets, it's essential to consider additional metrics such as accuracy, recall, F1-score, or confusion matrix.
- The F1-score is a popular metric for evaluating binary classifiers' effectiveness as it combines recall and precision into a single score, providing an overall measure of accuracy.

### Precision

- Precision in a classification or prediction model serves as a measure of its accuracy.
- It represents the ratio of true positives (correctly predicted positive cases) to the total number of positive cases predicted by the model.
- Put simply, precision calculates the percentage of predicted positive cases that are actually positive.
- High precision indicates the model effectively identifies positive cases with few false positives.
- Conversely, low precision suggests a higher occurrence of false positives in the model's output, potentially leading to inaccurate or misleading results.

### Recall

- **Recall as a Measure of Completeness**: 
  - The completeness of a categorization or prediction model is assessed through recall.
  - Recall is defined within the model's output as the ratio of true positives (correctly predicted positive cases) to the sum of true positives and false negatives (positive cases missed by the model).
  - Essentially, recall measures the percentage of actual positive cases accurately identified by the model.

- **High and Low Recall Significance**:
  - High recall indicates the model accurately recognizes most positive cases, minimizing overlook.
  - Conversely, low recall suggests a higher probability of missing positive cases within the model, potentially leading to inaccurate or misleading outcomes.

## 1. Logistic Regression

```python
#Logistic Regression parameters for K-fold cross vaidation:
params = {"C": [0.01, 0.1, 1, 10, 100, 1000]}
folds = KFold(n_splits = 5, shuffle = True, random_state = 4)


#Perform cross validation:
model_cv = GridSearchCV(estimator = LogisticRegression(),
                        param_grid = params, 
                        scoring= 'roc_auc', 
                        cv = folds, 
                        n_jobs=-1,
                        verbose = 1,
                        return_train_score=True) 

#Perform hyperparameter tuning:
model_cv.fit(X_train, y_train)

#Print the evaluation result by choosing a evaluation metric:
print('Best ROC AUC score: ', model_cv.best_score_)

#Print the optimum value of hyperparameters:
print('Best hyperparameters: ', model_cv.best_params_)
```

#### Prediction and model evalution on the train set

```python
#Creating function to display ROC-AUC score, f1 score and classification report:
def display_scores(y_test, y_pred):
    '''
    Display ROC-AUC score, f1 score and classification report of a model.
    '''
    print(f"F1 Score: {round(f1_score(y_test, y_pred)*100,2)}%") 
    print(f"Classification Report: \n {classification_report(y_test, y_pred)}")
```

```python
# ROC Curve function
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize = (8, 6))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
#Predicted probability:
y_train_pred_prob = log_reg_imb_model.predict_proba(X_train)[:,1]
```

# Strategic Selection of XGBoost Model for Credit Card Fraud Detection

- In our endeavor to combat credit card fraud, our foremost goal is to attain a high recall rate. This is essential for identifying a substantial portion of genuine fraudulent transactions, thereby establishing a robust defense against potential high-value fraudulent activities.

- Given the significant risks posed by fraudulent transactions, including monetary losses, damage to credibility, and erosion of trust for banking institutions, prioritizing recall emerges as a strategic imperative. This approach safeguards the financial stability of banks and upholds the interests of their customers.

# Model Evaluation and Selection Process

In our analysis, we methodically explored a range of machine learning models, harnessing the capabilities of the ADASYN (Adaptive Synthetic Sampling) technique to address the significant class imbalance in the dataset. Amidst the array of models evaluated, the XGBoost model distinguished itself as a top performer. Below, we outline the pivotal discoveries:

- **ROC Score:** The XGBoost model achieved an impressive ROC score of 99%, underscoring its proficiency in discerning between positive and negative classes effectively.

- **Recall Rate:** A notable highlight was the exceptionally high recall rate of 86% observed in the balanced dataset. This metric holds paramount importance in ensuring the accurate identification of a significant portion of actual fraudulent transactions.

# Rationale for Choosing XGBoost

- Following a thorough examination of performance metrics, the XGBoost model, in conjunction with the ADASYN technique applied to a balanced dataset, emerges as the optimal choice for our credit card fraud detection solution.
